```python
# Create a new branch called 'feature-branch'
!git checkout -b feature-branch  # Makes a new branch for features/experiments
python

# Add another function to perform division with error handling
def divide_numbers(a, b):
    """
    Divides a by b and returns the result. Handles division by zero gracefully.
    """
    if b != 0:
        return a / b
    else:
        return "Cannot divide by zero"
```

```python
# Print the result of dividing 10 by 2
print(divide_numbers(10, 2))  # Output: 5.0
python
# Add, commit, and push the updated notebook from your feature branch
!git add assignment2.ipynb  # Add changes
!git commit -m "Add division function"  # Commit the new feature
!git push --set-upstream origin feature-branch  # P
```
